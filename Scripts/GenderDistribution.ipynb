# 1. Import

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import csv
```

# 2. Get data

```python
noc_region_df = pd.read_csv("./Data/noc_regions.csv")
athlete_events_df = pd.read_csv("./Data/athlete_events.csv")
df = pd.merge(athlete_events_df, noc_region_df)
df = df.rename(columns={'region':'Region', 'notes':'Notes'}).drop_duplicates()
```

# 3. Processing

```python
df['Age'].fillna((df['Age'].mean()),inplace=True)
df['Height'].fillna((df['Height'].mean()),inplace=True)
df['Weight'].fillna((df['Weight'].mean()),inplace=True)
df['Region'].fillna(df['Team'],inplace=True)
df['Notes'].fillna('No Notes',inplace=True)
df['Medal'].fillna('No Medal',inplace =True)
```

# 4.1. Gender by Years

```python
def plot_gender_over_time(all_data_df):
    n_athletes_over_years = all_data_df.groupby(['Year', 'Sex'])['ID'].nunique().sort_index(ascending=False)
    plt.figure(figsize=(20,5))
    sns.lineplot(x=n_athletes_over_years.index.get_level_values('Year'), y=n_athletes_over_years.values, hue=n_athletes_over_years.index.get_level_values('Sex'), marker='o')
    plt.xlabel('Year', size=20)
    plt.ylabel('Number of athletes', size=20)
    plt.title('Number of athletes changed over time by genders', size=30)
    plt.xticks(list(range(1896, 2020))[::4])
    plt.xlim(min(all_data_df["Year"]),max(all_data_df["Year"]))
    plt.grid();
plot_gender_over_time(df)
```
